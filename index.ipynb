```python
def parNumeros(df, columna):
    
    pares = []
    
    for valor in df[columna]:
        if valor % 2 == 0:
            pares.append(valor)
    
    return pares
```

```python
fig1 = go.Figure()
fig1.add_trace(go.Bar(x = df['Frecuencia'],y = df['Categoria']))

fig1.show()
```
